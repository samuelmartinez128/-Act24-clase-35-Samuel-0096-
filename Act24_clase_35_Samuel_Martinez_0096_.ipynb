### **Inmueble**
#### - Atributos: direccion, metrosCuadrados, numeroHabitaciones, precio
#### - Funciones: mostrar(), negociarPrecio(), realizarContrato()

```python
# --- PROGRAMA POO: CLASE INMUEBLE ---

class Inmueble:
    # El método __init__ inicializa los atributos solicitados
    def __init__(self, direccion, metrosCuadrados, numeroHabitaciones, precio):
        self.direccion = direccion
        self.metrosCuadrados = metrosCuadrados
        self.numeroHabitaciones = numeroHabitaciones
        self.precio = precio

    # Función 1: Muestra la información del inmueble
    def mostrar(self):
        print(f"Inmueble en {self.direccion} | Tamaño: {self.metrosCuadrados}m² | Habitaciones: {self.numeroHabitaciones} | Precio: ${self.precio}")

    # Función 2: Simula la negociación del precio
    def negociarPrecio(self):
        print(f"Iniciando negociación para el inmueble en {self.direccion}. El precio actual es de ${self.precio} 💰.")

    # Función 3: Simula la realización del contrato
    def realizarContrato(self):
        print(f"Generando contrato legal para la propiedad de {self.metrosCuadrados} metros cuadrados... 📝✅")

# --- ENTRADA DE DATOS (Input) ---

print("--- Registro del Inmueble 1 ---")
dir1 = input("Dirección: ")
m2_1 = int(input("Metros cuadrados (solo números): "))
hab1 = int(input("Número de habitaciones: "))
pre1 = int(input("Precio: "))
inmueble1 = Inmueble(dir1, m2_1, hab1, pre1)

print("\n--- Registro del Inmueble 2 ---")
dir2 = input("Dirección: ")
m2_2 = int(input("Metros cuadrados (solo números): "))
hab2 = int(input("Número de habitaciones: "))
pre2 = int(input("Precio: "))
inmueble2 = Inmueble(dir2, m2_2, hab2, pre2)

# --- LLAMADAS A LAS FUNCIONES PARA CADA OBJETO ---

print("\n--- Acciones del Inmueble 1 ---")
inmueble1.mostrar()
inmueble1.negociarPrecio()
inmueble1.realizarContrato()

print("\n--- Acciones del Inmueble 2 ---")
inmueble2.mostrar()
inmueble2.negociarPrecio()
inmueble2.realizarContrato()

print("\nPrograma terminado")
print("Gestión Inmobiliaria - Adaptado del modelo de Samuel Martinez NL =35")
```
